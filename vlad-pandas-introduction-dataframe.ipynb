```python
df_lazarev = df[(df['performer'] == '–ª–∞–∑–∞—Ä–µ–≤') & (df['likeCount'] > 0)] #get the columns lazarev and likeCount
max_index = df_lazarev['likeCount'].idxmax() #get the index of the row that has the biggest number in likeCount
max_Lazarev_like = df.loc[max_index, 'likeCount'] #find that index in df and return value
print(f"Lazarev's maximum number of likes is: {max_Lazarev_like}")
```

```python
df_filtered = df[['title', 'viewCount']]
m_index = df_filtered['viewCount'].idxmax()
max_views_title = df.loc[m_index, 'title']
print(f"The title with maximum number of views is: {max_views_title}")
```

```python
df_eldjey = df[(df['performer'] == '—ç–ª–¥–∂–µ–π') & (df['likeCount'] > 0)]
df_eldjey_mean = df_eldjey['likeCount'].mean()
eldjey_titles_above_average = df[(df['performer'] == '—ç–ª–¥–∂–µ–π') & (df['likeCount'] > df_eldjey_mean)]
print(f"The performer 'eldjey' gets {df_eldjey_mean:.0f} likes on average.")
print("His titles that have likes above average are:")
for index, row in eldjey_titles_above_average.iterrows():
    print(f"{row['title']} - the number of likes: {row['likeCount']}.")
```
